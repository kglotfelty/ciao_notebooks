## Fill in sources

```bash
/bin/rm -rf srcfill
mkdir srcfill
```

```bash
roi \
  infile=14024/wav.out \
  fovregion="region(14024/repro/acisf14024_repro_fov1.fits)" \
  outsrcfile="srcfill/srcs_%04d.fits" \
  group=indi \
  radiusmode=mul bkgradius=3 \
  bkgfactor=2 bkgfunction=add \
  targetbkg=target clob+ mode=h 
```

```bash
/bin/ls srcfill | egrep -v srcs.lis > srcfill/srcs.lis
```

```bash
stk_build @srcfill/srcs.lis stdout
```

```bash
ds9 14024/a401_0.500-3.000_thresh.img -scale log \
  -block 4 \
  -view colorbar no -view magnifier no -view panner no -view info no -view buttons no &
  

sleep 5


stk_build @srcfill/srcs.lis stdout | xargs -I@ xpaset -p ds9 regions load @'[bkgreg]'

xpaset -p ds9 saveimage png ds9_02.png 
xpaset -p ds9 quit

display < ds9_02.png
```

```bash
splitroi @srcfill/srcs.lis points
```

### Run `dmfilth`

```bash
punlearn dmfilth
```

```bash
dmfilth \
  14024/a401_0.500-3.000_thresh.img \
  a401.img \
  method=POISSON \
  srclist=@points.src.reg \
  bkglist=@points.bg.reg \
  mode=h clob+
```

```bash
ds9 a401.img -scale log -block 4 \
  -view colorbar no -view magnifier no -view panner no -view info no -view buttons no \
  -saveimage png ds9_03.png -quit

display < ds9_03.png
```

```bash
aconvolve a401.img a401_sm.img lib:'gaus(2,5,1,2,2)' meth=slide clob+
```

```bash
aconvolve 14024/a401_0.500-3.000_thresh.expmap a401_sm.exp lib:'gaus(2,5,1,2,2)' meth=slide clob+
```

```bash
dmimgcalc a401_sm.img a401_sm.exp a401_sm.flux div clob+ 
```

```bash
dmimgthresh a401_sm.flux a401_sm_thresh.flux cut=5% exp=14024/a401_0.500-3.000_thresh.expmap value=0 clob+
```

```bash
ds9 a401_sm_thresh.flux -scale log -block 4 \
  -view colorbar no -view magnifier no -view panner no -view info no -view buttons no \
  -saveimage png ds9_03a.png -quit

display < ds9_03a.png
```

## Now run `dmellipse`

```bash
punlearn dmellipse

dmellipse a401_sm_thresh.flux'[bin sky=2]' a401.ellipses fraction="lgrid(0.025:1.0:0.025)" \
  step=200 norm+ clob+ mode=h 
```

```bash
ds9 a401_sm_thresh.flux -region a401.ellipses -scale log \
  -block 4 \
  -view colorbar no -view magnifier no -view panner no -view info no -view buttons no \
  -saveimage png ds9_04.png -quit

display < ds9_04.png
```

## Now make model

```bash
dmstat a401_sm_thresh.flux cen- sig- verb=0
pget dmstat out_sum
```

```bash
plist emodel
```

```bash
dmcopy a401_sm_thresh.flux'[sky=region(14024/repro/acisf14024_repro_fov1.fits[ccd_id=0:3])][opt full]' a401_fov.img clob+
```

```bash
emodel a401.ellipses a401_fov.img a401_model.img \
  norm=0.010210171321 \
  weight=flat \
  clob+
  
  
```

```bash
ds9 a401_model.img -scale log \
  -block 4 \
  -view colorbar no -view magnifier no -view panner no -view info no -view buttons no \
  -saveimage png ds9_05.png -quit

display < ds9_05.png
```

```bash
dmimgcalc a401_sm_thresh.flux a401_model.img a401_resdiuals.img sub clob+
```

```bash
ds9 a401_resdiuals.img  \
  -block 4 -linear -scale mode 99 -smooth \
  -view colorbar no -view magnifier no -view panner no -view info no -view buttons no \
  -saveimage png ds9_06.png -quit

display < ds9_06.png  
```
